```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python

df = pd.read_csv('/content/breast-cancer.csv')

df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
x = df.drop(columns = ['id', 'diagnosis'], axis = 1)
y = df['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
print(x.shape, x_train.shape, x_test.shape)
```

```python
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(x_train, y_train)
```

```python
test_data_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(y_test, test_data_prediction)
print('Accuracy score of the test data: ', test_data_accuracy)
```

```python
metrics = {
    'Accuracy': accuracy_score(y_test, test_data_prediction),
    'Precision': precision_score(y_test, test_data_prediction),
    'Recall': recall_score(y_test, test_data_prediction),
    'F1 Score': f1_score(y_test, test_data_prediction)
}

df_metrics = pd.DataFrame(metrics, index=['Score'])

conf_matrix = confusion_matrix(y_test, test_data_prediction)
print('Confusion Matrix: \n', conf_matrix)

print('\nClassification Metrics:')
print(df_metrics)
```
